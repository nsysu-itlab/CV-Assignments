```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from scipy import linalg
%matplotlib inline

def show_gradient_descent(A, b, init_condition=np.array([[-2, 2]]), verbose=True):
#TODO:  enter your code here, and remove the following line (30%)
   x = init_condition.T
   alpha = 0.02
   max_iter = 1000
   logs = []
   logs.append(x)
   f = np.linalg.norm((np.dot(A, x)-b))**2
   for k in range(max_iter):
      if verbose==True:
         print(f"{k:2d}, ({float(x[0]):.4f}, {float(x[1]):.4f}), {f:8f}")
      x_prev = x
      grad_f = 2*((A.T.dot(A).dot(x))-(A.T.dot(b)))
      x = x - alpha * grad_f
      logs.append(x)
      if np.linalg.norm(x-x_prev) <= 1e-8:
         break
      else:
         f = np.linalg.norm(A.dot(x)-b)**2
   logs = np.array(logs).squeeze()
   return logs
```

```python
def draw_gradient_descent(A, b, logs1, logs2):
  xx, yy = np.meshgrid(np.arange(-2,2.1,0.1), np.arange(-2,2.1,0.1))
  F1 = A[0][0]*xx + A[0][1]*yy - b[0];
  F2 = A[1][0]*xx + A[1][1]*yy - b[1];
  zz = F1*F1 + F2*F2;
  plt3d = plt.figure().add_subplot(projection='3d')
  plt3d.plot_surface(xx, yy, zz, alpha=0.7, cmap=cm.coolwarm) 

  plt3d.plot(logs1[:,0], logs1[:,1], 'r.')
  # plot line eq_m and eq_b is calculate from two minimum
  eq_m = 0.49875 
  eq_b = -0.7492
  x = np.linspace(-2, 2, 100)
  # plt3d.plot([logs1[-1,0], logs2[-1, 0]], [logs1[-1, 1], logs2[-1, 1]], 'b')
  plt3d.plot(x, eq_m*x+eq_b, 'b')
  pass
```

```python
if __name__ == '__main__':
  b = np.array([[1], [2]])
  A = np.array([[1, -2], [1, -2]])
  logs = show_gradient_descent(A, b)
  other_min = show_gradient_descent(A, b, np.array([[0,0]]), verbose=False)
  draw_gradient_descent(A, b, logs, other_min)
```
